```python
#IMPORT LIBRARY YANG DIBUTUHKAN
import numpy as np
import pandas as pd
```

```python
#Menambahkan dataset
#Data yang digunakan adalah data belanja Konsumen dimana terdapat invoice id, branch, city, dll

#load dataset menggunakan URL
df = pd.read_csv("https://raw.githubusercontent.com/Crocellz/Supermarket-sales/main/supermarket_sales%20-%20Sheet1.csv")
```

```python
#Menampilkan 5 record teratas yang telah di load, menggunakan head()

df.head()
```

```python
#Melihat informasi dataset dari kolom Age (200 data terdapat 192 data yang non null) berarti 8 data lainnya bernilai null/NaN
df.info()
```

```python
#Mendeskripsikan kolom yang memiliki missing value yaitu "Quantity"

df['Quantity'].describe()
```

```python
#Terlihat bahwa terdapat 8 missing value pada kolom Quantity

df.isnull().sum()
```

```python
#Mengganti Missing value menggunakan Metode Imputation (mean)

mean = df["Quantity"].mean()
df["Quantity"] = df["Quantity"].replace(np.nan, mean)
df.head()
```

```python
#Mengganti Missing value menggunakan Metode Imputation (Mode)

df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mode()[0])
df.head()
```

```python
#Mengganti Missing value menggunakan Metode Imputation (median)

median = df["Quantity"].median()
df["Quantity"] = df["Quantity"].replace(np.nan, median)
df.head()
```

```python
#Panjang data 

print("\nPanjang Data :" , df.shape)
```

```python
#Menghapus Missing Value 

df.dropna(inplace=True)
print(df.isnull().sum())
#setelah dihapus Missing Value pada kolom age menjadi 0
```
